## In this notebook

- We implement  ~~spacy entities and noun phrase based n-grams extraction~~ Textacy to pick out key terms and phrases from reviews




```python
import pandas as pd
```

### load data

```python
df = pd.read_excel("../data/test_data.xlsx", index_col=0)
df['contents'] = df['title'] + '.\n\n' + df['review']
```

### tokenization

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
df['review_tokens'] = df['contents'].apply(lambda x: [t.text.lower() for t in nlp(x)])
```

### extract keyterms

```python
import textacy
```

```python
from textacy import extract
```

```python
from textacy.extract import keyterms
```

```python
# construct a massive spacy doc for the keyterms module
massive_doc = '\n\n\n'.join(df['contents'].tolist() )
```

```python
massive_spacy_doc = nlp(massive_doc)
```

```python
textrank_df = pd.DataFrame.from_dict(dict(keyterms.textrank(massive_spacy_doc, normalize='lemma', topn=250)),
                    orient='index').reset_index().rename(columns={'index': 'phrase', 0:'score'})
```

```python
textrank_df
```

```python
textrank_df[ ~textrank_df['phrase'].str.contains(r'(Sony|Bravia|LG)', case=False)].head(25)
```

```python
scake_df = pd.DataFrame.from_dict(dict(keyterms.scake(massive_spacy_doc, normalize='lemma', topn=100)),
                    orient='index').reset_index().rename(columns={'index': 'phrase', 0:'score'}) 
                    
```

```python
scake_df[ ~scake_df['phrase'].str.contains(r'(Sony|Bravia|LG)', case=False)].head(25)
```

```python
yake_df = pd.DataFrame.from_dict(dict(keyterms.yake(massive_spacy_doc, normalize='lemma', topn=100)),
                    orient='index').reset_index().rename(columns={'index': 'phrase', 0:'score'}) 
```

```python
yake_df[ ~yake_df['phrase'].str.contains(r'(Sony|Bravia|LG)', case=False)].head(25)
```
